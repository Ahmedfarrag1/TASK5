# importing the needed libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

# loading the dataframe using pandas read_CSV function

```python
df = pd.read_csv('/content/drive/MyDrive/shrink_ray_dataset.csv')

df.shape
```

# Extract the knob values (X) and the shrinkage values (y) from the DataFrame

```python
X = df.iloc[:, 0].values.reshape(-1, 1)  # Assuming the first column is the input feature
y = df.iloc[:, 1].values  # Assuming the second column is the target variable
```

# create a linear regression instance and fit to the data

```python
model = LinearRegression()
model.fit(X, y)
```

# making predictions


```python
y_prediction = model.predict(X)
```

# Visualize the data and the linear regression line attaching the scatterplot

```python
plt.scatter(X, y, color='blue')
plt.plot(X, y_prediction, color='red')
plt.xlabel('Knob')
plt.ylabel('shrinckage')
plt.title('Linear Regression')
plt.show()
```

# method of calculating the corresponding knob value to 85% shrinkage

```python
# Get the model's coefficients and intercept
slope = model.coef_[0]
intercept = model.intercept_
shrink_percentage = 85

# Solve the equation y = mx + c for x
knob = (shrink_percentage - intercept) / slope

print(f"The knob value corresponding to {shrink_percentage}% shrink = {knob}")
```
